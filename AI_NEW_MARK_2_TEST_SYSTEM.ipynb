```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from tkinter import *
import numpy as np
```

```python
data_train = pd.read_csv("data_train.csv",dtype={'colx': int})
data_train[['year', 'month', 'day']] = data_train['data'].str.split(' ', expand=True).astype(int)
data_train.head()
```

```python
day = data_train[["year","month","day"]].values
six = data_train[["colx 6"]].values
first_3 = data_train[["colx 12"]].values
first_3 = first_3.astype(int)

two = data_train[["colx 9"]].values
two.astype(int)

first_3 = first_3 / np.max(first_3)
two = two / np.max(two)
```

```python
def model_LinearRegression(d, m, y, data1, data2):
    model = LinearRegression()
    model.fit(data1, data2)
    predict = model.predict([[d, m, y]])
    return predict[0]

# Example usage
iD = int(input("Day: "))
iM = int(input("Month: "))
iY = int(input("Year: "))
sur = str(input("this code is demo so dataset and code have a some bug. you want to run : yes or no"))
if sur == "yes":
    print("\nเลขท้าย 6 ตัว: %06d" % model_LinearRegression(iD, iM, iY, day, six)[0])
    print("เลขหน้า 3 ตัว: %03d" % model_LinearRegression(iD, iM, iY, day, first_3)[0])
    print("เลขท้าย 2 ตัว: %06d" % model_LinearRegression(iD, iM, iY, day, two)[0])
else :
    print("ok")


# Assuming 'day', 'main_train', and other datasets are correctly preprocessed

```

```python

file_path = 'data_train.csv'

data = pd.read_csv(file_path)
data.head()
# Convert 'converted_date' to separate numerical columns
data[['year', 'month', 'day']] = data['data'].str.split(' ', expand=True).astype(int)

# Drop the original 'converted_date' column
data = data.drop(columns=['data'])

# Extract the relevant features
main_train = data[["colx 4"]].astype(float)
first_3 = data[["colx 11"]].astype(float)
last_3 = data[["colx 12"]].astype(float)
last_2 = data[["colx 9"]].astype(float)

# Function to predict using Linear Regression
def model_LinearRegression(d, m, y, data1, data2):
    model = LinearRegression()
    model.fit(data1, data2)
    predict = model.predict([[d, m, y]])
    return predict[0]

# Example usage
iD = int(input("Day: "))
iM = int(input("Month: "))
iY = int(input("Year: "))

# Assuming 'day', 'main_train', and other datasets are correctly preprocessed
print("\nเลขท้าย 6 ตัว: %06d" % model_LinearRegression(iD, iM, iY, main_train, last_3)[0])
print("เลขหน้า 3 ตัว: %03d" % model_LinearRegression(iD, iM, iY, main_train, first_3)[0])
print("เลขท้าย 6 ตัว: %06d" % model_LinearRegression(iD, iM, iY, main_train, last_3)[0])
print("เลขท้าย 3 ตัว: %03d" % model_LinearRegression(iD, iM, iY, main_train, last_3)[0])
print("เลขท้าย 3 ตัว 2: %03d" % model_LinearRegression(iD, iM, iY, main_train, last_2)[0])
```
